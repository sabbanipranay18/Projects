```python
import pandas as pd
import numpy as np
import json
```

# Loading "match_summary.json" file

```python
with open('D:/Portfolio Project/Best 11 World cup/t20_json_files/t20_json_files/t20_wc_match_results.json') as data:
    df = json.load(data)
df
```

Converting data into "DataFrame" so that we can manipulate the data

```python
df_match_summary = pd.DataFrame(df[0]['matchSummary'])
df_match_summary
```

Lets Rename "ScoreCard" to "MatchID", lets treat this column as primary key which is later used to connect to other tables...

```python
df_match_summary.rename({"scorecard" : "matchID"}, axis = 1, inplace = True)
df_match_summary.head(10)
```

Exporting this "match_summary" to CSV format

```python
match_id_dict = {}
for index, row in df_match_summary.iterrows():
    key1 = row["team1"] + " Vs " + row["team2"]
    key2 = row["team2"] + " Vs " + row["team1"]
    match_id_dict[key1] = row["matchID"] 
    match_id_dict[key2] = row["matchID"]
```

Exporting this "match_summary" table data into CSV format

```python
df_match_summary.to_csv("D:/Portfolio Project/Best 11 World cup/T20_CSV/match_summary_data.csv", index = False)
```

# Loading "batting_summary.json" file

```python
with open("D:/Portfolio Project/Best 11 World cup/t20_json_files/t20_json_files/t20_wc_batting_summary.json") as data2:
    df2 = json.load(data2)
df2
```

Making all records from different arrays into one list

```python
df_batting = []
for i in df2:
    df_batting.extend(i["battingSummary"])
df_batting
```

Now converting this data into "DataFrame"

```python
df_batting_summary = pd.DataFrame(df_batting)
df_batting_summary
```

In the Dismissal column we just need "out/not_out", to get that we can write a function saying if we have any data then it is out, else not_out.

```python
df_batting_summary["out/not_out"] = df_batting_summary["dismissal"].apply(lambda x : "out" if len(x) > 0 else "not_out")
df_batting_summary.head(10)
```

After adding a new column for "out/not_out", we dont need dismissal column so i removed that column 

```python
df_batting_summary.drop(columns=["dismissal"], inplace = True)
```

```python
df_batting_summary.head(20)
```

We created a Dictionary where the match names is stored with unique match id of that, here i created a two combination keys and appended the unique key called "matchID" to that particular name.

```python
match_id_dict = {}
for index, row in df_match_summary.iterrows():
    key1 = row["team1"] + " Vs " + row["team2"]
    key2 = row["team2"] + " Vs " + row["team1"]
    match_id_dict[key1] = row["matchID"] 
    match_id_dict[key2] = row["matchID"]
match_id_dict
```

Here we mapped the unique "matchID" to team names using match_id_dict

```python
df_batting_summary["matchID"] = df_batting_summary["match"].map(match_id_dict)
df_batting_summary.head(10)
```

Removing unnecessary special-characters from table

```python
df_batting_summary["batsmanName"] = df_batting_summary["batsmanName"].apply(lambda x: x.replace("â€", " "))
df_batting_summary.head(10)
```

Exporting this "batting_summary" table data into "CSV" format

```python
df_batting_summary.to_csv("D:/Portfolio Project/Best 11 World cup/T20_CSV/batting_summary_data.csv", index = False)
```
